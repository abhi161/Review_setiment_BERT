```python
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
import re
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")

model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
sent =pipeline("sentiment-analysis",model= model, tokenizer=tokenizer)
```

```python
result=sent("very bad product quality but design is worst")
```

```python
result[0]['label']
```

```python
URL= "https://www.yelp.com/biz/taco-bell-daly-city-4"

review = requests.get(URL)
soup = BeautifulSoup(review.text,"html.parser")
regex= re.compile('.*comment.*')
r =soup.find_all("p",{'class':regex})
rev = [t.text for t in r]
rev
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.array(rev),columns=["reviews"])
df
```

```python
df["rating"] = df.reviews.apply(lambda x:sent(x)[0]["label"])
```
